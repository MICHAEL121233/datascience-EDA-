```python
days=[1,2,3,4,5]

sleeping=[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m',label='sleeping',lw=5)
plt.plot([],[],color='c',label='eating',lw=5)
plt.plot([],[],color='r',label='working',lw=5)
plt.plot([],[],color='k',label='playing',lw=5)

plt.stackplot(days,sleeping,eating,working,playing, colors=['m','c','r','k'])
plt.title('Stack or Area plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
# THE AGGFUNC IS USED TO SPECIFY THE STAT ANALYSIS YOU WANT, BY DEFAULT IT IS MEAN

pivot3= pd.pivot_table(data, index='Region', values='Sales', aggfunc='sum')
pivot3
```

```python
# THE AGGFUNC CAN TAKE A LIST OF STAT ANALYSIS TO PERFORM
pivot4= pd.pivot_table(data, index='Region', values='Sales', aggfunc=['sum','mean'])
pivot4
```

```python
# THE COLUMNS ARGUMENT IS USED TO SELECT THE COLUMN YOU WANT TO PERFORM THE PIVOT OPS
pivot5= pd.pivot_table(data, index='Region', values='Sales', aggfunc= 'sum',
                      columns='Type')
pivot5
```

```python
# MARGINS IS USED TO PERFORM THE TOTAL ROW AND COLUMN SUMMATION

pivot6= pd.pivot_table(data, index='Region', values='Sales', aggfunc= 'sum',
                      columns='Type', margins=True)
pivot6
```

```python
# THE MARGINS NAME IS USED TO RENAME THE MARINGS COLUMN
pivot7= pd.pivot_table(data, index='Region', values='Sales', aggfunc= 'sum',
                      columns='Type', margins=True, margins_name='Total')
pivot7
```

# Data Scaling

```python
# data scaling is simply making your programme understand your data by using
# same scale for all your data
# for example, if a = 1000m and b = 20 kilometer, 
# you and i know that 20km is longer than 1000m, but your programme does not 
# hence
a = 1000 #meter
b = 20   #km
b>a

# To address situations like this is why we have data scaling
```

```python
# MinMaxScaler
# x_new = (x_old - min)/(max-min)
```

```python
import numpy as np
a = np.array([10,20,30,40,50])
b = np.array([100,200,300,400,500])
```

```python
# this function is an example of how the MinMaxScaler works

def minmax(ar):
    ar_new = (ar-ar.min())/(ar.max()-ar.min())
    return ar_new
```

```python
minmax(a)
```

```python
minmax(b)
```
